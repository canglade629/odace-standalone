```sql
%python
csv_path = "/Volumes/odace/map/files/geo/communes-france-2025.csv"

# Read the CSV file into a Spark DataFrame
df_communes = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(csv_path)
)

# Write the DataFrame to the Delta table
df_communes.write.format("delta").option("mergeSchema", "true").mode("overwrite").saveAsTable("odace.map.bronze_geo")
```
