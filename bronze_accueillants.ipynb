```python
import pandas as pd

excel_path = "/Volumes/odace/map/files/accueillants/accueillants.xlsx"

# Read the Excel file: skip first 5 rows, row 6 is header, data starts at row 7
df_pd = pd.read_excel(
    excel_path,
    header=0     # header is now the first row after skipping
)

# Rename columns to replace spaces with underscores
new_columns = {col: col.replace(' ', '_') for col in df_pd.columns}
df_pd.rename(columns=new_columns, inplace=True)

# Convert to Spark DataFrame
df_spark = spark.createDataFrame(df_pd)

# Create or replace the Delta table in odace.map schema
df_spark.write.format("delta").mode("overwrite").saveAsTable("odace.map.bronze_accueillants")
```
